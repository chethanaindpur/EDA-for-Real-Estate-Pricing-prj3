# Univariate Analysis: Price Distribution (Histogram + KDE Plot)

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["SalePrice"], bins=30, kde=True, color="blue")
plt.title("Histogram & KDE of SalePrice")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
num_variables = ['LotArea', 'MasVnrArea', '1stFlrSF', 'GrLivArea']
fig, axes = plt.subplots(1, len(num_variables), figsize=(6 * len(num_variables), 5))
for i, var in enumerate(num_variables):
    sns.histplot(data=df, x=var, kde=True, ax=axes[i], color="blue")
    axes[i].set_title(f'Distribution of {var}', fontsize=14, fontweight='bold') 
    axes[i].set_xlabel(var, fontsize=12, fontweight='bold') 
    axes[i].set_ylabel("Frequency", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
cat_variables = ['LandSlope', 'Street', 'LotConfig', 'LotShape']
fig, axes = plt.subplots(1, len(cat_variables), figsize=(4 * len(cat_variables), 5))
for i, var in enumerate(cat_variables):
    sns.countplot(data=df, x=var, ax=axes[i], palette="pastel")
    axes[i].set_title(f'Count of {var}')
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
cat_variables = ['LandSlope', 'Street', 'LotConfig', 'LotShape']
fig, axes = plt.subplots(1, len(cat_variables), figsize=(4 * len(cat_variables), 5))
for i, var in enumerate(cat_variables):
    category_counts = df[var].value_counts() 
    axes[i].pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', 
                colors=sns.color_palette("pastel"), startangle=90)
    axes[i].set_title(f"Distribution of {var}")
plt.tight_layout() 
plt.show()
```

# Feature Engineering for House Price Prediction

# Creating New Features

```python
#combining existing features.
#Total Bathrooms
df["TotalBath"] = df["FullBath"] + (0.5 * df["HalfBath"]) + df["BsmtFullBath"] + (0.5 * df["BsmtHalfBath"])
df
```

```python
#Total House Size (Above and Basement)
df["TotalSF"] = df["TotalBsmtSF"] + df["GrLivArea"]
df
```

```python
#Age of the House at the Time of Sale
df["HouseAge"] = df["YrSold"] - df["YearBuilt"]
df
```

# Remodeling Indicator

```python
df["Remodeled"] = (df["YearBuilt"] != df["YearRemodAdd"]).astype(int)
df
```

#  Transformation of Skewed Features

```python
# Apply log transformation to handle skewness
df["Log_SalePrice"] = np.log1p(df["SalePrice"])  
df["Log_LotArea"] = np.log1p(df["LotArea"]) 
df["Log_TotalSF"] = np.log1p(df["TotalSF"])  

# Display transformed data
df[["Log_SalePrice", "Log_LotArea", "Log_TotalSF"]].head()
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Log-Transformed SalePrice
sns.histplot(df["Log_SalePrice"], bins=50, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Log-Transformed SalePrice")
axes[0].set_xlabel("Log Sale Price")
axes[0].set_ylabel("Frequency")

# Log-Transformed LotArea
sns.histplot(df["Log_LotArea"], bins=50, kde=True, color="green", ax=axes[1])
axes[1].set_title("Log-Transformed LotArea")
axes[1].set_xlabel("Log Lot Area")
axes[1].set_ylabel("Frequency")

# Log-Transformed TotalSF
sns.histplot(df["Log_TotalSF"], bins=50, kde=True, color="red", ax=axes[2])
axes[2].set_title("Log-Transformed TotalSF")
axes[2].set_xlabel("Log TotalSF")
axes[2].set_ylabel("Frequency")

# Adjusted layout
plt.tight_layout()
plt.show()
```

# Feature Interaction (Multiplying Features)

```python
#capture hidden relationships between variables.
df["Qual_LivingArea"] = df["OverallQual"] * df["GrLivArea"]
df["GarageQual_Area"] = df["GarageCars"] * df["GarageArea"]
df
```

# Feature Selection (Keeping the Best Features)

```python
# list of columns to remove
cols_to_remove = ['LotFrontage', 'BedroomAbvGr', 'ScreenPorch', '3SsnPorch', 'PoolArea', 
                  'BsmtFinSF2', 'MiscVal', 'BsmtHalfBath', 'LowQualFinSF', 'OverallCond', 
                  'EnclosedPorch', 'KitchenAbvGr', '2ndFlrSF']

# Checking the cols which are existing in dataframes
existing_cols_to_remove = [col for col in cols_to_remove if col in df.columns]

# Dropping only existing columns
df.drop(existing_cols_to_remove, axis=1, inplace=True)

print("Remaining columns:", df.columns)
```

# Feature Engineering and Size Impact on House Prices

```python
# size-related features
size_features = ["SalePrice", "TotalBsmtSF", "1stFlrSF", "GrLivArea", "GarageArea", "LotArea"]
corr_matrix = df[size_features].corr()

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation of Size Features with SalePrice")
plt.show()
```

```python
# Total square footage including basement
df["TotalSF"] = df["TotalBsmtSF"] + df["1stFlrSF"]

# Total number of bathrooms including half-baths
df["TotalBathrooms"] = df["FullBath"] + (0.5 * df["HalfBath"]) + df["BsmtFullBath"]

# Total Porch area
df["TotalPorchSF"] = df["OpenPorchSF"]

# Garage Size Efficiency-size per car space
df["GarageEfficiency"] = df["GarageArea"] / (df["GarageCars"] + 1)
```

```python
# Log transformation skewed features
df["Log_TotalSF"] = np.log1p(df["TotalSF"])
df["Log_TotalBathrooms"] = np.log1p(df["TotalBathrooms"])
df["Log_TotalPorchSF"] = np.log1p(df["TotalPorchSF"])

# Visualize transformed data
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.histplot(df["Log_TotalSF"], bins=50, kde=True, color="blue", ax=axes[0])
sns.histplot(df["Log_TotalBathrooms"], bins=50, kde=True, color="green", ax=axes[1])
sns.histplot(df["Log_TotalPorchSF"], bins=50, kde=True, color="red", ax=axes[2])
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))

sns.regplot(x="TotalSF", y="SalePrice", data=df, scatter_kws={'alpha':0.5}, ax=axes[0, 0])
axes[0, 0].set_title("TotalSF vs SalePrice")

sns.regplot(x="TotalBathrooms", y="SalePrice", data=df, scatter_kws={'alpha':0.5}, ax=axes[0, 1])
axes[0, 1].set_title("TotalBathrooms vs SalePrice")

sns.regplot(x="GarageEfficiency", y="SalePrice", data=df, scatter_kws={'alpha':0.5}, ax=axes[1, 0])
axes[1, 0].set_title("GarageEfficiency vs SalePrice")

sns.regplot(x="TotalPorchSF", y="SalePrice", data=df, scatter_kws={'alpha':0.5}, ax=axes[1, 1])
axes[1, 1].set_title("TotalPorchSF vs SalePrice")

plt.tight_layout()
plt.show()
```
